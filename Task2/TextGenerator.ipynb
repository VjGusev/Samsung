#### Задание: 
Требуется написать программу-генератор текста.

Английского, но лучше русского, а еще лучше - русского художественного.

Для решения задачи генерации текста будем использовать рекуррентную нейронную сеть.
Основная особенность такой сети заключается в том, что она подстраивается под особенности текущей последовательности объектов и дальше может предсказывать какой объект выдать следущим, на основе анализа текущего объекта.
Осуществляет она предсказания за счет скрытого вектора h и обученных весов.

Задачи:
1. Собрать данные, которые характеризуют нашу задачу. Например, если задача генерации английских текстов, тогда английские тексты, а если русских, тогда русские тексты.
2. Обучить на них рекуррентную нейронную сеть. Произойдет настройка весовых матриц.
3. Дальше можно генерировать текст.

Дальше можно пробовать улучшать алгоритм, например, пробовать улавливать n-граммы. Это поможет уловить нам стилистику текста.

можем попробовать уйти совсем в другую сторону и пробовать тематическое моделирование.
